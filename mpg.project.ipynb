### Task 1

Use seaborn.load_dataset https://seaborn.pydata.org/generated/seaborn.load_dataset.html to load the mpg (miles per gallon, a measure of fuel economy) data

```python
import seaborn as sns
```

```python
sns.load_dataset("mpg")
mpg = sns.load_dataset("mpg")
```

### Task 2

Create a visual showing the relationship between horsepower and fuel economy

```python
cars = mpg.groupby("name").agg({"mpg":"sum","horsepower":"sum"})
sns.relplot(data=cars,x="mpg", y="horsepower")
```

### Task 3

Create a table summarizing fuel economy by country of origin

```python
mpg.groupby(by="origin")["mpg"].mean()
```

### Task 4

Create a visual showing how fuel economy has changed over time

```python
sns.lineplot(data=mpg,x="model_year", y="mpg")
```

### Task 5

Create a visual showing the relationship between cylinder and fuel economy

```python
sns.catplot(data=mpg,x='cylinders',y='mpg',kind='bar',errorbar=None)
```
